Simple linear regression is a statistical method used to model the relationship between two continuous variables. It aims to find the linear relationship between the independent variable (predictor) and the dependent variable (response).

In this type of regression, we have a single independent variable that predicts the outcome of a dependent variable. The relationship between the two variables is represented by a straight line equation of the form:

y=mx+b

Where:

1. y is the dependent variable (response or target variable)
2. x is the independent variable (predictor)
3. m is the slope of the line (the change in y for a unit change in x)
4. b is the y-intercept (the value of y when x is 0)

The main objectives of simple linear regression are:

1. Modeling: 
To describe and quantify the relationship between two variables.
Prediction: To predict the value of the dependent variable based on the value of the independent variable.
The process of simple linear regression involves:

2. Data Collection: 
Collecting pairs of observations of the independent and dependent variables.

3. Model Building: 
Fitting a linear equation to the data using methods like the least squares approach, minimizing the sum of the squared differences between the observed and predicted values.

4. Model Evaluation:
Assessing the goodness of fit of the model using metrics like R-squared, Mean Squared Error (MSE), etc., to understand how well the model explains the variability in the data.

5. Prediction:
Using the fitted model to predict the dependent variable for new or unseen values of the independent variable.

Assumptions of simple linear regression include:

1. There is a linear relationship between the independent and dependent variables.
2. Residuals (the differences between observed and predicted values) are normally distributed.
3. Homoscedasticity: Residuals have constant variance across all levels of the independent variable.
4. Independence: Observations are independent of each other.

Simple linear regression serves as a foundational technique in statistical modeling and is often a starting point for more complex regression analyses. It's commonly used in various fields like economics, finance, social sciences, and machine learning for predictive modeling.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Generating random data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 3 + 4 * X + np.random.randn(100, 1)
```

```python
X
```

```python
y
```

```python
# Creating and fitting the regression model
model = LinearRegression()
model.fit(X, y)

# Predicting the values using the model
y_pred = model.predict(X)

# Calculating R-squared score
r_squared = r2_score(y, y_pred)
print(f'R-squared score: {r_squared:.4f}')
```

```python
# Scatter plot before regression
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Actual data')
plt.title('Scatter Plot Before Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Scatter plot after regression
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Predicted data')
plt.title('Scatter Plot After Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```
